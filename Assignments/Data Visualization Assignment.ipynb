## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x = np.linspace(0,10,5000)
plt.xlabel('X')
plt.ylabel("Sin(X)")
plt.title("Sine Wave graph for X")
plt.plot(x,np.sin(x),linestyle='-.',color='k')
plt.show()
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
d = {'Store':['Delhi','Mumbai','Chennai','Kolkata','Heyderabad'],
     'Clothing':[82,78,48,46,39],'Equipment':[48,31,51,29,18],
    'Accessories':[15,20,14,13,9]}
d = pd.DataFrame(d)
d
```

```python
width = 0.25
plt.bar(d.index,d.Clothing,width,label='Clothong',color='m')
plt.bar(d.index+width,d.Equipment,width,label='Equipment',color='b')
plt.bar(d.index+2*width,d.Accessories,width,label='Accessories',color='r')
plt.legend()
plt.xlabel('Stores')
plt.ylabel('Scores')
plt.xticks(d.index,list(d.Store))
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.barplot(x='class',y='survived',hue='class',palette='Accent_r',data=titanic)
plt.legend(loc='upper left')
plt.show()
```
